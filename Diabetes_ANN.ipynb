```python
import csv
from array_solution import *
from ClassANN import *
```

```python
with open('pima-indians-diabetes-clean.csv') as Data_file:
    Data_list = csv.reader(Data_file)
    Data = [Data for Data in Data_list]
    Rawdata_array = Array(len(Data))
    for i in range(len(Data)):
        Rawdata_array[i] = Data[i]
```

```python
#ใส่ข้อมูลอาเรย์ของRaw Data
ANN = Ann(Rawdata_array)
```

```python
#ลอง train 1 fold
#round 1 fold 1,2,3,4,5,6 เป็น train set และ fold7 เป็น test set
ANN.fit(300,0.01,7,None)
```
